```python
import requests
import json
```

```python
#Grab API Endpoint
resp = requests.get("https://opendata.maryland.gov/resource/jwfa-fdxs.json?$limit=1200")
resp.status_code
```

```python
resp.headers['content-type']
```

```python
resp_obj = resp.json()
resp_obj
```

```python
#Write data to new file
with open('crime-data.json', 'w') as f:
    json.dump(resp_obj, f)
```
